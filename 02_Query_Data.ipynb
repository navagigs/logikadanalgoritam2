# Query data dengan index

## Mengambil satu atau beberapa baris tertentu dengan index berupa integer

```python
# mengambil 1 (satu) baris tertentu
df.iloc[1]
```

```python
# mengambil baris index no 22 dan Nama Lengkap
df.loc[1,'Nama Lengkap'] #angka berdasarkan nomor index dan nama kolom
```

```python
# mengambil beberapa baris tertentu dengan index berupa list
df.iloc[[2,22,44]] #angka berdasarkan nomor index
```

```python
# baris ke-3, kolom ke-5
df.iloc[2,4] #output 'Kab. Pekalongan'
```

```python
# baris ke-3,23, dan 45, kolom ke-4 dab 5
df.iloc[[2,22,44], [3,4]] 
```

```python
#baris ke-6 hingga 9, kolom ke-1 hingga 2
df.iloc[5:9, 1:2] 
```

```python
df.iloc[:10, 1:] #tanpa angka di depan range berarti semua index sebelumnya, tanpa angka di belakang range berarti semua index setelahnya
```

```python
df.iloc[-1, -1] #Ambil 1 baris/kolom terakhir saja
```

```python
df.iloc[-20:, -5:] #20 baris terakhir, 5 kolom terakhir
```

```python
df.loc[[40,45,50],['Nama Lengkap','Kota/Kabupaten']]
```

```python
df[(df['Provinsi']=='Jawa Barat')] #mengambil data berdasarkan provinsi
```

```python
df.loc[df['Provinsi' ] == 'Jawa Barat' , 'Provinsi' ]
```

```python
df[(df['Kota/Kabupaten'].str.contains('Kota Bandung')) & (df['Provinsi']=='Jawa Barat')]
```

```python
df[df['Kota/Kabupaten'].str.contains('Bekasi')]
```

```python
df[df['Kota/Kabupaten'].str.contains('Bekasi') & df['JK'].str.contains('L')]
```
